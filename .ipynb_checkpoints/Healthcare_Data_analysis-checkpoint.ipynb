```python
# Create Age group 
def age_group(age):
    if age < 18:
        return '<18'
    elif 18 <= age < 30:
        return '18-29'
    elif 30 <= age < 45:
        return '30-44'
    elif 45 <= age < 60:
        return '45-59'
    else:
        return '60+'
```

```python
# Apply the function to create the 'age_group' column
df['age_group'] = df['age'].apply(age_group)
```

#### Relationships and interactions between 'age_group' and 'medical_condition'

```python
# crosstab age_group and medical_condition
crosstab_result4 = pd.crosstab(df['age_group'], df['medical_condition'])
round(crosstab_result4,2)
```

Those under 18 are at 23.28% in Obesity situation followed by 18.10% Arthritis and Cancer 18-29 years: Obesity (17.02%) 30-44 years: Arthritis (17.14%) 45-59 years: Obesity (17.11%) 60+ years: almost equivalent proportion in medical situations


```python
# Chi-square test for 'admission_type' and 'medical_condition'
chi2_stats, chi2_pvalue, _, _ = chi2_contingency(crosstab_result4)
print(f'chi2_pvalue: {chi2_pvalue}')
```

Interpretation 
( p )-value >0,05: we don't reject the null hypothesis.  
This implies that there's not a significant association between age_group and medical_condition in the dataset.

```python
# crosstab test_result and admission_type
crosstab_result6 = pd.crosstab(df['test_results'], df['medication'])
round(crosstab_result6,2)
```

```python
crosstab_result6.plot(kind='bar',rot=0, figsize=(6, 4))

plt.title('Patient Distribution by test_results and Medication', fontsize=12)
plt.xlabel('Test resultat', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Medication', fontsize=10)
plt.tight_layout()

plt.show()
```

#### Testing correlation between 'test_results' and 'medication'

The null hypothesis of the Chi-square test is that the two categorical variables are independent. 

```python
chi2_stats, chi2_pvalue, _, _ = chi2_contingency(crosstab_result6)
print(f'chi2_pvalue: {chi2_pvalue}')
```

Interpretation 
( p )-value >0,05: we don't reject the null hypothesis.  
This implies that there's not a significant association between test_result and medication in the dataset.

#### Calculating the Pearson correlation coefficient between billing_amount and stay_lengh to quantify their linear relationship

```python
# Calculate Pearson coefficient
coef_pearson= df['billing_amount'].corr(df['lengh_stay'])
print(f"pearson correlation coefficient: {coef_pearson}")
```

Pearson coefficient close to 0 means there is almost no linear relationship between the two variables. The slight negative value (-0.0049) suggests a negligible tendency for billing_amount to decrease as stay_length increases, but this effect is statistically insignificant

#### Admission Seasonality

```python
monthly_admissions = df['date_of_admission'].dt.month.value_counts().sort_index()

monthly_admissions_df = pd.DataFrame({'Month': monthly_admissions.index, 'Admissions': monthly_admissions.values})

monthly_admissions_df['Month'] = monthly_admissions_df['Month'].apply(lambda x: 
    pd.to_datetime(str(x), format='%m').strftime('%B'))  # Convert month numbers to names
plt.figure(figsize=(8, 5))

sns.lineplot(data=monthly_admissions_df, x='Month', y='Admissions', marker='o')

plt.title('Monthly Admissions Trend')
plt.xlabel('Month')
plt.ylabel('Number of Admissions')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
